Regression Assignment


1) What is Simple Linear Regression?
-> Simple linear regression is a statistical method that models the relationship between a single dependent variable (y) and one independent predictor variable (x) using a straight line.

2) What are the key assumptions of Simple Linear Regression?
-> Simple linear regression relies on key assumptions for valid, reliable results: a linear relationship between variables, independent observations, homoscedasticity (constant variance of residuals), and normally distributed residuals.

3) What does the coefficient m represent in the equation Y=mx+c?
-> In the equation y=mx+c, 𝑚 represents the slope or gradient.It defines the steepness and direction of the line.

4) What does the intercept c represent in the equation Y=mx+c?
-> In the linear equation y=mx+c, the intercept c represents the y-intercept, which is the specific point where the straight line crosses the y-axis. It indicates the value of y when
x= 0, representing the starting vertical position of the line on a graph.

5) How do we calculate the slope m in simple linear regression?
-> The formula to calculate the slope (m) of the best fit line in linear regression is obtained by dividing the sum of the products of the differences between the observed values and their means by the sum of the squared differences between the independent variable values and their mean.

6) What is the purpose of the least squares method in simple linear regression?
-> The purpose of the least squares method in simple linear regression is to find the line of best fit for a set of data points by minimizing the sum of the squared vertical residuals (errors). It calculates the optimal slope and intercept, reducing the difference between observed and predicted values.

7) How is the coefficient of determination (R^2) interpreted in simple linear regression?
-> In simple linear regression, the coefficient of determination (R^2) is interpreted as the proportion of the total variation in the dependent variable (Y) that is explained by the independent variable through the linear relationship.

8) What is Multiple Linear Regression?
-> Multiple Linear Regression (MLR) is a statistical technique that models the relationship between one continuous dependent variable and two or more independent (predictor) variables by fitting a linear equation to observed data.

9) What is the main difference between simple and multiple linear regression?
-> Simple linear regression models the relationship between a dependent variable and one independent variable using a straight line (y=mx+c).
Multiple linear regression uses two or more independent variables to predict the dependent variable, allowing for more complex, nuanced, and accurate predictive modeling (y=m1x1+m2x2+....mnxn+c).

10) What are the key assumptions of Multiple Linear Regression?
-> The key assumptions of multiple linear regression are linearity between predictors and the dependent variable, independence of observations, homoscedasticity (constant variance of residuals), normality of error distribution, and little to no multicollinearity among independent variables. These ensure unbiased, reliable estimates.

11) What is heteroscedasticity and how does it affect the results of a multiple linear regression model?
-> Heteroscedasticity occurs when the variance of error terms (residuals) is not constant across all levels of the independent variables in a regression model. Instead of a uniform scatter, residuals show a "fan" or "cone" shape pattern. This violation of Ordinary Least Squares (OLS) assumptions means that while coefficients remain unbiased, they are no longer efficient, causing unreliable standard errors, invalid confidence intervals, and flawed hypothesis tests.

# Impact
* Inefficient Estimates: OLS estimators are no longer the Best Linear Unbiased Estimators (BLUE); they do not have the minimum variance.
* Unreliable Standard Errors: The standard errors are biased, which means they do not accurately reflect the precision of the coefficient estimates.

12) How can u improve a multiple linear regression model with high multicollinearity?
-> You can improve a multiple linear regression model with high multicollinearity by removing highly correlated predictors, using dimensionality reduction techniques like Principal Component Analysis (PCA), or employing regularization methods such as Ridge and Lasso regression.

13) What are some common techniques for transforming categorical variables for use in regression models?
-> Common techniques for transforming categorical variables in regression include One-Hot Encoding (creating binary columns for nominal data), Label/Ordinal Encoding (assigning integers to ordered data), and Dummy Coding (using k-1 variables to avoid collinearity). Other methods include Binary Encoding for high cardinality and Target Encoding.

14) What is the role of interaction terms in multiple linear regression?
-> nteraction terms in multiple linear regression model the combined effect of two or more independent variables on a dependent variable, allowing the impact of one predictor to change depending on the level of another.

# Roles
* Capturing Synergies or Moderation: They test if the relationship between the outcome and a predictor varies at different values of another variable (moderator).
* Modeling Non-parallel Slopes: In graphical terms, while main effects assume parallel slopes, interaction terms allow the slope of the regression line to differ between groups.
* Improving Model Fit: Including these terms can result in a better fit to the data and higher predictive performance.

15) How can the interpretation of intercept differ between Simple and multiple linear regression?
-> The intercept represents the dependent variable's (y) estimated value when all independent variables (x) are zero. In simple regression, it is the average y at x=0.  In multiple regression, it is the baseline (y) when all predictors are held constant at zero, often indicating adjusted, conditional effects.

16) What is the significance of slope in regression analysis, and how does it affect predictions?
-> Significance of the Slope in Regression:
* Direction of Relationship: A positive slope indicates that as x increases, y increases. A negative slope indicates that as x increases, y decreases.
* Rate of Change: The slope quantifies the exact, average rate of change. For example, if the slope of studying hours vs. test scores is 5, it means each extra hour of study is predicted to raise the score by 5 points.
* Strength of Relationship: A steeper slope (larger absolute value) indicates a stronger, more sensitive relationship between the variables, while a flatter slope indicates a weaker relationship.

How the Slope Affects Predictions:
* Linear Estimation: The slope is used to calculate the predicted value of y (y=mx+c). If the slope is high, small changes in x cause large changes in the predicted y.
* Prediction Sensitivity: The magnitude of the slope tells you how sensitive the dependent variable is to changes in the independent variable.* Accuracy: A well-calculated, statistically significant slope reduces prediction error by correctly identifying the trend in the data.

17) How does the intercept in a regression model provide context for the relationship between variables?
-> The intercept in a regression model provides crucial context by representing the expected value of the dependent variable (y) when all independent variables (x) are zero. It serves as the baseline or anchor point for the entire regression line, dictating where the relationship begins on the y-axis.

18) What are the limitations of using R^2 as a sole measure of model performance?
-> Using R^2 as a sole measure of model performance is limited because it always increases when adding more variables, even if they are irrelevant, potentially causing overfitting. It measures correlation, not accuracy, meaning a high R^2 can hide biased models, non-linear relationships, or poor predictions.

19) How would you interpret a large standard error for a regression coefficient?
-> A large standard error for a regression coefficient indicates low precision and high uncertainty in the estimate, meaning the coefficient is likely to vary significantly across different samples. This suggests the independent variable has a weak, inconsistent, or imprecise relationship with the dependent variable, often resulting in statistically insignificant results.

20) How can heteroscadasticity be identified in residual plots and why is it important to address it?
-> Heteroscedasticity is identified in residual plots by a non-constant spread of residuals, typically appearing as a fan, cone, or megaphone shape where residual variance increases or decreases with fitted values. Addressing it is crucial because it violates Ordinary Least Squares (OLS) assumptions, leading to inefficient estimates, biased standard errors, and unreliable hypothesis tests.

21) What does it mean if a multiple linear regression model has a high R^2 but low adjusted R^2?
-> A high R^2 paired with a low adjusted R^2 in multiple linear regression signifies that the model includes too many predictors—particularly irrelevant or redundant ones—relative to the number of data points. It suggests the model is overfitted, indicating it fits the training data well but will perform poorly on new data.

22) Why is it important to scale variables in Multiple Linear Regression?
-> Scaling variables in Multiple Linear Regression is crucial to ensure all predictors contribute equally to the model, especially when they are on different scales (e.g., age vs. income). It prevents variables with larger magnitudes from dominating, aids faster convergence for gradient-based optimizations, and allows for direct comparison of coefficient importance.

23) What is polynomial regression?
-> Polynomial regression is a type of machine learning regression analysis that models the relationship between a dependent variable and an independent variable as an n^th degree polynomial. It is used to fit non-linear data by creating a curved line, making it more flexible than simple linear regression when data points do not follow a straight line.

24) How does polynomial regression differ from linear regression?
-> Linear Regression fits a straight line to data, ideal for linear trends, while Polynomial Regression uses curves to model complex, non-linear relationships. Linear is simpler, faster, and less prone to overfitting, whereas polynomial offers higher flexibility at the risk of overfitting.

25) When is polynomial regression used?
-> Polynomial regression is used when data shows a non-linear relationship (e.g., curves, parabolas) that simple linear regression cannot accurately capture. It is best applied to model complex trends, such as increasing growth rates, or when residuals from a linear model display a pattern.

28) What are the limitations of polynomial regression?
-> Overfitting: High-degree polynomials can fit noise in the training data rather than the underlying trend, leading to poor generalization on new data.
Poor Extrapolation: The model behaves unpredictably outside the range of observed data, often resulting in erratic predictions at the edges (extrapolation)
Computational Complexity: Higher-degree models increase the computational cost and require more resources, especially with large datasets.

29) What methods can be used to evaluate model fit when selecting the degree of a polynomial?
-> Primary methods include k-fold cross-validation, analyzing residuals plots, and comparing information criteria (AIC/BIC) or using forward/backward selection to test coefficient significance.

30) Why is visualization important in polynomial regression?
-> Visualization is critical in polynomial regression to detect non-linear patterns, verify if the chosen degree fits the data correctly, and prevent overfitting (too complex) or underfitting (too simple). It helps determine if the model accurately captures trends rather than just noise, allowing for easier interpretation of complex curves compared to analyzing raw numerical coefficients.
